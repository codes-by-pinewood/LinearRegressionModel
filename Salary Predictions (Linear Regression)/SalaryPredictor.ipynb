```python
# Import dataset

file_path = "Data Source/Salary_Data.csv"
df = pd.read_csv(file_path)
```

```python
df.describe()
```

### Graphical depiction

```python
#Graph size and title
plt.figure(figsize = [8, 5])
plt.title("Years of experience vs Salary ")

#Scatterplot x,y coordinates
sns.scatterplot(x = df["YearsExperience"], y=df["Salary"])

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

```python
#Linear Regression Plot
plt.figure(figsize = [8,5])
plt.title("Regression Line Distribution of Years of Experience vs Salary")

sns.regplot(x = df["YearsExperience"], y=df["Salary"])

plt.xlabel("Years of Experience")
```

#### As expected an increase in years of experience correlates to a higher salary.
